```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('householdtask3.csv')
data.head(10)
```

```
#scatter plot
plt.scatter(data['year'],data['own'])
#adding the title
plt.title("Year Over Own Scatter plot")
#adding x and y label
plt.xlabel("Year")
plt.ylabel("Own")
plt.colorbar()

#showing
plt.show()

```

```
#Line chart
plt.plot(data['year'],data['own'])
#adding the title
plt.title("Year Over Own Line chart")
#adding x and y label
plt.xlabel("Year")
plt.ylabel("Own")

#showing
plt.show()
#
```

```
# bar chart or bar plot
plt.bar(data['year'],data['own'])
#adding the title
plt.title("Year Over Own Bar chart")
#adding x and y label
plt.xlabel("Year")
plt.ylabel("Own")

#showing
plt.show()
#
```

```
#histogram of income
plt.hist(data['income'])
#adding the title
plt.title("Income Histogram")
#adding x and y label
plt.xlabel("Income")
plt.ylabel("Frequency")

#showing
plt.show()
#
```
